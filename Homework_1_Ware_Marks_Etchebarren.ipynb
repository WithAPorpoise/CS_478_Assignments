Activation Function : Identity

g(x) = x

```python
#Imports
import numpy as np
import matplotlib.pyplot as plt
#Defining x inputs
x = np.arange(-5, 5, 0.01)
# Identity function
y = x 
#Visualizing the graph
plt.plot(x,y) 
plt.title('Identity')
plt.xlabel('x')
#Use label vertical axis g(x) instead of y to plot
plt.ylabel('g(x)');
#Add a background grid to the plot
plt.grid()
#Show plot
plt.show()
```

Activation Function : Sigmoid

g(x) = sigmoid.JPG

```python

import numpy as np
import matplotlib.pyplot as plt
x = np.arange(-5, 5, 0.01)
# @TODO Signmoid activation function

e = np.e

y = (1 /(1+ (e**(-x))))

plt.plot(x,y) 
plt.title('Sigmoid')
plt.xlabel('x')
plt.ylabel('g(x)');
plt.grid()
plt.show()
```

Activation Function : Tahn

g(x) = tahn.JPG

```python

import numpy as np
import matplotlib.pyplot as plt
x = np.arange(-5, 5, 0.01)
# Tanh activation function
# 
e = np.e

y = ((e**x) - (e**(-x))) / ((e**x) +(e**(-x)))
plt.plot(x,y) 
plt.title('Hypertangent')
plt.xlabel('x')
plt.ylabel('g(x)');
plt.grid()
plt.show()
```

Activation Function : Relu

g(x) = Relu.JPG

```python

import numpy as np
import matplotlib.pyplot as plt
x = np.arange(-5, 5, 0.01)
# ReLU activation function

y = []
for i in x:
  if i <= 0:
    y.append(0)
  else:
    y.append(i)


plt.plot(x, y)
plt.title('Rectified Linear Unit')
plt.xlabel('x')
plt.ylabel('g(x)');
plt.grid()
plt.show()
```

Activation Function : Gelu

g(x) = GELU.JPG

```python
import math
import numpy as np
import matplotlib.pyplot as plt
x = np.arange(-5, 5, 0.01)
# GeLU activation function

def gelu(x):
  return (0.5*x * (1 + math.erf(x/math.sqrt(2))))

GeLU = np.vectorize(gelu)

plt.plot(x,GeLU(x)) 
plt.title('Gaussian Error Linear Unit')
plt.xlabel('x')
plt.ylabel('g(x)');
plt.grid()
plt.show()
```

Activation Function : Elu

let alpha = 1

g(x) = elu.JPG

```python

import numpy as np
import matplotlib.pyplot as plt
x = np.arange(-5, 5, 0.01)
# ELU activation function

e=np.e

def elu(x,a):
  return max(a *((e**x) -1), x)

eLU = np.vectorize(elu)
plt.plot(x, eLU(x,1))
plt.title('Exponential Linear Unit')
plt.xlabel('x')
plt.ylabel('g(x)');
plt.grid()
plt.show()
```

Activation Function : SELU

g(x) = SELU.JPG

```python

import numpy as np
import matplotlib.pyplot as plt
x = np.arange(-5, 5, 0.01)
# SELU activation function

y = []
e=np.e
l = 1.0507
a = 1.67326

def selu(x):
  if x >= 0:
    return l * x
  else:
    return (l* a *((e**x) -1))

SeLU = np.vectorize(selu)
plt.plot(x, SeLU(x))
plt.title('Scaled Exponential Linear Unit')
plt.xlabel('x')
plt.ylabel('g(x)');
plt.grid()
plt.show()
```

Activation Function : Leaky ReLU

g(x) = LRelu.JPG

```python

import numpy as np
import matplotlib.pyplot as plt
x = np.arange(-5, 5, 0.01)
# Leaky ReLU
 
def relu(x, a):
  return max(a * x, x)

LeakyReLU = np.vectorize(relu)
plt.plot(x, LeakyReLU(x, .01))
plt.title('Leaky Rectified Linear Unit')
plt.xlabel('x')
plt.ylabel('g(x)');
plt.grid()
plt.show()
```

```python

import numpy as np
import matplotlib.pyplot as plt
x = np.arange(-5, 5, 0.01)
# PReLU activation function
 
def relu(x, a):
  return max(a * x, x)

PReLU = np.vectorize(relu)
plt.plot(x, PReLU(x, .1))
plt.title('Parametric Rectified Linear Unit')
plt.xlabel('x')
plt.ylabel('g(x)');
plt.grid()
plt.show()
```

Activation Function : Gaussian

g(x) = Gaussian.JPG

```python

import numpy as np
import matplotlib.pyplot as plt
x = np.arange(-5, 5, 0.01)
# Gausian activation function

e = np.e
y = e**(-x**2)

plt.plot(x,y) 
plt.title('Gaussian')
plt.xlabel('x')
plt.ylabel('g(x)');
plt.grid()
plt.show()
```
